# **Age Group Distribution**

````python
# This code performs age analysis and compares the age distribution of customers by default status

# Define the age groups and their corresponding bins
ages = (df['AGE'])
bins = [20, 30, 40, 50, 60, 70, 80]
group_names = ['20s', '30s', '40s', '50s', '60s', '70s']

# Create a new DataFrame for ages and their corresponding age group
ages_df = pd.DataFrame({'ages': ages})
ages_df['age_group'] = pd.cut(ages_df['ages'], bins, labels=group_names, right=False)

# Count the number of customers in each age group
age_group_counts = ages_df['age_group'].value_counts()

# Calculate the percentage of customers in each age group
total_count = len(ages)
percentage = (age_group_counts / total_count) * 100

# Plot a bar chart to show the distribution of customers by age group
sns.barplot(x=age_group_counts.index, y=percentage)
plt.ylabel('Percentage')
plt.xlabel('Age Group')
plt.title('Distribution of Customers by Age Group')
plt.show()


# Compare the age distributions of customers who defaulted and those who didn't
age_not_default = (df[df["default"]==0]["AGE"])
age_default = (df[df['default']==1]["AGE"])

# Create new DataFrames for ages and their corresponding age group for customers who defaulted and those who didn't
ages_not_default_df = pd.DataFrame({'ages': age_not_default})
ages_default_df = pd.DataFrame({'ages': age_default})

ages_not_default_df['age_group'] = pd.cut(ages_not_default_df['ages'], bins, labels=group_names, right=False)
ages_default_df['age_group'] = pd.cut(ages_default_df['ages'], bins, labels=group_names, right=False)

# Count the number of customers who defaulted and those who didn't in each age group
age_group_not_default_counts = ages_not_default_df['age_group'].value_counts()
age_group_default_counts = ages_default_df['age_group'].value_counts()

# Calculate the percentage of customers who defaulted and those who didn't in each age group
total_not_default_count = len(age_not_default)
total_default_count = len(age_default)

percentage_not_default = (age_group_not_default_counts / total_not_default_count) * 100
percentage_default = (age_group_default_counts / total_default_count) * 100

# Plot a line chart to compare the age distributions of customers who defaulted and those who didn't
sns.lineplot(x=percentage_not_default.index, y=percentage_not_default, color='green')
sns.lineplot(x=percentage_default.index, y=percentage_default, color='red')
plt.ylabel('Percentage')
plt.xlabel('Age Group')
plt.title('Comparison of Age Distributions by Default Status')
plt.show()
````

# Monthly Payment Status Counts.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the figure
plt.figure(figsize=(20,10))

# Specify the columns to plot
pay_col = ['PAY_APR', 'PAY_MAY','PAY_JUN','PAY_JUL','PAY_AUG','PAY_SEPT']

# Loop through each column and plot a countplot
for num, column in enumerate(pay_col):
    # Create a subplot
    plt.subplot(2,3,num+1)
    
    # Get the unique values and their counts
    counts = df[column].value_counts()
    
    # Plot a bar plot with the counts
    sns.barplot(x=counts.index, y=counts.values)
    
    # Set the title of the plot
    plt.title(f'{column.title()}',weight='bold')
    
    # Set the layout
    plt.tight_layout()
    
# Show the plot
plt.show()
````

The num variable keeps track of the iteration number, and the plt.subplot() function is used to place each countplot in a subplot within the figure. The subplot is specified by the first two arguments (2, 3), which specify the number of rows and columns respectively, and the third argument (num + 1), which specifies the subplot number.

The plt.tight_layout() function is used to adjust the subplots so that they fit well within the figure.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

1- Null Hypothesis - There is no relation between Categorical Variables and Default

Alternate Hypothesis - There is a relationship between Categorical Variables and Default

#### 2. Perform an appropriate statistical test.

````python
from scipy.stats import chi2_contingency

def hypothesis_test_chi2(df, categorical_variable, alpha=0.05):
    '''
    Conduct a hypothesis test using chi-square test for independence between
    the categorical variable and the default status of the loan

    Args:
    df (pandas.DataFrame): The dataset
    categorical_variable (str): The name of the categorical variable to be tested
    alpha (float): The significance level for the test

    Returns:
    str: A string indicating whether to reject or fail to reject the null hypothesis
    '''

    # Split the data into default and non-default groups
    default = df[df['default'] == 1]
    non_default = df[df['default'] == 0]

    # Create a contingency table of the categorical variable and default
    cont = pd.crosstab(df['default'], df[categorical_variable])

    # Conduct a chi-square test for the independence of the categorical variable and default
    chi2, p_value, dof, expected = chi2_contingency(cont)

    # Make a decision based on the p-value and alpha
    if p_value < alpha:
        return f"Reject the null hypothesis. There is a significant association between {categorical_variable} and default."
    else:
        return f"Fail to reject the null hypothesis. There is no significant association between {categorical_variable} and default."
````


**The function performs a chi-square test for independence between a categorical variable and the default status of a loan. It returns whether to reject or fail to reject the null hypothesis based on a specified significance level.**

Code: The contingency table is created using the pd.crosstab function 

It tests the null hypothesis that the two variables are independent against the alternative hypothesis that they are dependent
 
The test calculates the chi-square statistic, degrees of freedom (dof), expected frequencies under the null hypothesis, and the p-value.

The results of the chi-square test are stored in the variables chi2, p_value, dof, and expected.

````python
# Define a list of categorical variables
categorical = ['SEX', 'EDUCATION', 'MARRIAGE','PAY_SEPT', 'PAY_AUG', 'PAY_JUL', 'PAY_JUN', 'PAY_MAY', 'PAY_APR']

# Loop through the list of categorical variables
for categorical_variable in categorical:
    # Perform a chi-square test for the independence of the categorical variable and default
    result = hypothesis_test_chi2(df, categorical_variable)

    # Print out the result of the test
    print(result)
````


**The code performs a chi-square test for independence between each categorical variable in the list and the default status of a loan, using the hypothesis_test_chi2 function. It prints out the result of each test to indicate whether to reject or fail to reject the null hypothesis.**

# **Capping**

**Capping Outliers using IQR (Interquartile Range) method is a technique for handling extreme values (outliers) in a dataset**. In this method, the lower and upper limits are defined as Q1 - kIQR and Q3 + kIQR respectively, where Q1 and Q3 are the first and third quartiles, IQR is the interquartile range, and k is a constant typically set to 1.5. Any value below the lower limit or above the upper limit is capped (replaced) with the respective limit value. This helps to reduce the effect of extreme values on the analysis of the dataset.

````python
#Capping Outlier using IQR method 
for col in numeric_cols:
    upper = df_copy[col].quantile(0.75) + 1.5 * (df_copy[col].quantile(0.75) - df_copy[col].quantile(0.25))
    df_copy[col] = np.where(df_copy[col] > upper, upper, df_copy[col])
    print("{}: {}".format(col, outliers))
````

# function for evaluation metrics

````python
def evaluation_metrics(model, X_train, y_train, X_test, y_test):
    model.fit(X_train, y_train)
    'function for evaluation metrics'
    
    # Predict classes for train and test data
    train_class_preds = model.predict(X_train)
    test_class_preds = model.predict(X_test)
    
    # Calculate the evaluation metrics
    train_accuracy = accuracy_score(y_train, train_class_preds)
    test_accuracy = accuracy_score(y_test, test_class_preds)
    train_precision = precision_score(y_train, train_class_preds)
    test_precision = precision_score(y_test, test_class_preds)
    train_recall = recall_score(y_train, train_class_preds)
    test_recall = recall_score(y_test, test_class_preds)
    train_f1 = f1_score(y_train, train_class_preds)
    test_f1 = f1_score(y_test, test_class_preds)
    train_roc_auc = roc_auc_score(y_train, train_class_preds)
    test_roc_auc = roc_auc_score(y_test, test_class_preds)
    
    # Store the evaluation metrics in a dictionary
    evaluation_metrics =OrderedDict([
    ('Train Accuracy: {:.2f}%'.format(train_accuracy*100), None),
    ('Test Accuracy: {:.2f}%'.format(test_accuracy*100), None),
    ('Train Precision: {:.2f}%'.format(train_precision*100), None),
    ('Test Precision: {:.2f}%'.format(test_precision*100), None),
    ('Train Recall: {:.2f}%'.format(train_recall*100), None),
    ('Test Recall: {:.2f}%'.format(test_recall*100), None),
    ('Train F1 Score: {:.2f}%'.format(train_f1*100), None),
    ('Test F1 Score: {:.2f}%'.format(test_f1*100), None),
    ('Train ROC-AUC: {:.2f}%'.format(train_roc_auc*100), None),
    ('Test ROC-AUC: {:.2f}%'.format(test_roc_auc*100), None)

])
    
    return evaluation_metrics
````

 The predict_proba method of this object is used to make predictions for new data

The predict_proba method returns an array of probabilities, one for each class, and the probabilities are stored in the train_preds and test_preds arrays.

'Train Accuracy': It is the ratio of correct predictions to the total number of predictions made on the training dataset. A higher accuracy indicates that the model is better at making correct predictions on the training data.

'Test Accuracy': It is the ratio of correct predictions to the total number of predictions made on the test dataset. A higher accuracy indicates that the model is better at generalizing to unseen data.

'Train Precision': It is the ratio of true positive predictions to the sum of true positive and false positive predictions. Precision measures the model's ability to correctly identify positive instances.

'Test Precision': It is the ratio of true positive predictions to the sum of true positive and false positive predictions, evaluated on the test dataset.

'Train Recall': It is the ratio of true positive predictions to the sum of true positive and false negative predictions. Recall measures the model's ability to find all positive instances.

'Test Recall': It is the ratio of true positive predictions to the sum of true positive and false negative predictions, evaluated on the test dataset.

'Train F1 Score': It is the harmonic mean of precision and recall. It provides a single number that balances both precision and recall.

'Test F1 Score': It is the harmonic mean of precision and recall, evaluated on the test dataset.

'Train ROC-AUC': It is the area under the Receiver Operating Characteristic (ROC) curve. It is a graphical representation of the performance of a binary classifier. The ROC-AUC is a single number that summarizes the overall performance of a classifier.

'Test ROC-AUC': It is the area under the Receiver Operating Characteristic (ROC) curve, evaluated on the test dataset.
